```python
# Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading test and training datasets

df1 = pd.read_csv('housing_train.csv')
df2 = pd.read_csv('housing_test.csv')
```

```python
# Observing what our data looks like

df1.head()
```

```python
# Observing the size of our training and test datasets

print("Training data:", df1.shape)
print("Test data:", df2.shape)
```

```python
# Checking data for skewness

sns.distplot(df1['SalePrice'])
```

```python
# Checking Skewness and kurtosis values to determine if the skewness is in acceptable range

print("Skewness: %f" % df1['SalePrice'].skew())
print("Kurtosis: %f" % df1['SalePrice'].kurt())
```

```python
# Acceptable value for skewness and kurtosis is usually considered to be Â± 1. Thus, we will fix the skew:

from scipy.stats import norm

df1['SalePrice'] = np.log1p(df1['SalePrice'])
sns.distplot(df1['SalePrice'], fit = norm);
```

```python
# Combining the test and training datasets, so that data processing is easier. 

Y_train = df1['SalePrice']
df1 = df1.drop('Id', axis=1)
df2 = df2.drop('Id', axis = 1)
train_test = pd.concat([df1, df2], sort = False, axis = 0)
```

```python
# We are going to be observing the percentage of missing values, so that we can drop columns with too many missing values

sum = train_test.isnull().sum().sort_values(ascending = False)

percentage = (train_test.isnull().sum() / train_test.isnull().count() * 100)
percentage = percentage.sort_values(ascending = False)

null = pd.concat([sum, percentage], axis = 1, keys = ['Total missing values', 'Percentage of missing values (%)'])

null.head(20)
```

```python
# Based on the percentages above, we are going to be dropping the following features from our data:

train_test = train_test.drop(['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'LotFrontage'], axis = 1)
```

```python
# Checking what we are left with

train_test.isnull().sum().sort_values(ascending = False).head(15)
```

```python
# We will fill the missing places with appropriate values

# To make our work easier, we have categorized features according to their common initial letters and formed separate loops for numerical and categorical variables

garage = [column for column in train_test if column.startswith('Garage')]
train_test[garage]

for i in train_test[garage].select_dtypes(exclude='object').columns:
    train_test[i] = train_test[i].fillna(0)

for i in train_test[garage].select_dtypes(include='object').columns:
    train_test[i] = train_test[i].fillna(train_test[i].mode()[0])

    
    
bsmt = [column for column in train_test if column.startswith('Bsmt')]

for i in train_test[bsmt].select_dtypes(exclude = 'object').columns:
    train_test[i] = train_test[i].fillna(0)

for i in train_test[bsmt].select_dtypes(include = 'object').columns:
     train_test[i] = train_test[i].fillna(train_test[i].mode()[0])
    
    
    
mas = [column for column in train_test if column.startswith('Mas')]

for i in train_test[mas].select_dtypes(exclude='object').columns:
    train_test[i] = train_test[i].fillna(0)

for i in train_test[mas].select_dtypes(include='object').columns:
     train_test[i] = train_test[i].fillna(train_test[i].mode()[0])
        
        

train_test['Electrical'] = train_test['Electrical'].fillna(train_test['Electrical'].mode()[0])
```

```python
# checking for any remaining missing values

print("Missing values remaining:", train_test.isnull().sum().sum())
```

```python
# We are going to be checking the skewness of numerical variables in the dataset

from scipy.stats import skew

numerical = train_test.dtypes[train_test.dtypes != 'object'].index
skew_features = train_test[numerical].apply(lambda x: skew(x)).sort_values(ascending=False)


# We are displaying features with skew Â± 0.5

skew = skew_features[abs(skew_features) > 0.5]

skew
```

```python
# We are fixing the skewness of the variables

# We are not removing outliers by conventional method because it causes a considerable loss of data and we want to retain as much data as possible

for column in skew.index:
    train_test[column] = np.log1p(train_test[column])
```

```python
# We have retained all our instances, while removing some features and cleaning our data

train_test.shape
```

```python
# We are going to be converting categorical data to numerical by creating dummy variables

train_test = pd.get_dummies(train_test)
train = pd.get_dummies(train)
train_test.head()
```

```python
# Importing libraries

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
k_means_df = train.copy().drop(['SalePrice'],axis=1)
df_PCA = train.copy().drop(['SalePrice'],axis=1)
```

```python
# Scaling the data

scaler = StandardScaler()
scaler.fit(df_PCA)
scaled_data = scaler.transform(df_PCA)
```

```python
# KMeans clustering

kmeans = KMeans(n_clusters = 2, random_state = 1)

kmeans = kmeans.fit(scaled_data)

print('The clusters are:', kmeans.labels_)

print('The inertia is: ',kmeans.inertia_)
```

```python
# Finding optimum number of clusters

no_cluster = [i for i in range(2, 21)]

inertia = []

for i in no_cluster:
    
    kmeans1 = KMeans(n_clusters = i, random_state = 1)
    
    kmeans1 = kmeans1.fit(k_means_df)
    
    iner = kmeans1.inertia_
    
    inertia.append(iner)
    
    print("The inertia for {} clusters is {}".format(i, iner))
```

```python
# Visualizing the inertia score against number of clusters

fig,(ax1)= plt.subplots(1, figsize = (13,5))

i = np.arange(len(no_cluster))

ax1.plot(i, inertia)
ax1.set_xticks(i)
ax1.set_xticklabels(no_cluster, rotation = 'horizontal')
plt.xlabel('Number of Clusters')
plt.ylabel('Score for Inertia')
```

```python
# With reference to the graph above, we will choose number of clusters as 7.

# The clusters are less internally coherent with less inertia and vice versa.

# Thus, we will choose number of clusters at which the rate of decrease of inertia score becomes relatively stable. 


kmeans1 = KMeans(n_clusters = 7, random_state = 5)

kmeans1=kmeans1.fit(k_means_df)

kmeans1.labels_

predictions = kmeans1.predict(k_means_df)

unique, counts = np.unique(predictions, return_counts = True)

counts = counts.reshape (1, 7)

counts_df = pd.DataFrame(counts, columns=['Cluster 1','Cluster 2','Cluster 3','Cluster 4','Cluster 5','Cluster 6','Cluster 7'])
counts_df
```

```python
# Importing libraries

from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
# Function for RMSE values

def rmse(ytrue, ypredicted):
    return np.sqrt(mean_squared_error(ytrue, ypredicted))
```

LASSO REGRESSION

```python
# Lasso regression using 5-fold cross validation

lasso = Lasso(max_iter = 100000, normalize = True)

lassocv = LassoCV(cv = 5, max_iter = 100000, normalize = True)

lassocv.fit(X_train, Y_train)

lasso.set_params(alpha = lassocv.alpha_)

lasso.fit(X_train, Y_train)

print("Alpha =", lassocv.alpha_)

print("RMSE =", rmse(Y_test, lasso.predict(X_test)))
```

RIDGE REGRESSION

```python
# Ridge regression using 5-fold cross validation

ridgecv = RidgeCV(cv = 5, normalize = True, scoring = 'neg_mean_squared_error')

ridgecv.fit(X_train, Y_train)

ridge = Ridge(alpha = ridgecv.alpha_, normalize = True)

ridge.fit(X_train, Y_train)

print("Alpha =", ridgecv.alpha_)

print("RMSE =", rmse(Y_test, ridge.predict(X_test)))
```

SUPPORT VECTOR REGRESSION

```python
# Support Vector regression using 5-fold cross validation

# Finding best parameters

from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
from sklearn.model_selection import KFold

cv = KFold(random_state = 1234, shuffle = True, n_splits = 5)

X_test_scale = RobustScaler().fit_transform(X_test)

X_train_scale = RobustScaler().fit_transform(X_train)

para = {'C':[20, 30, 40], 'epsilon':[0.1, 0.01, 0.05], 'gamma': [1e-4, 3e-4, 5e-4]}

svr = SVR(kernel = 'rbf')

clf = GridSearchCV(svr, para, cv = cv)

clf.fit(X_train_scale, Y_train)

clf.best_params_
```

```python
svr = SVR(kernel ='rbf', C = 40, epsilon = 0.1, gamma = 0.0005)

svr.fit(X_train_scale, Y_train)

print("RMSE =", rmse(Y_test, svr.predict(X_test_scale)))
```

```python
# Visualizing and comparing the RMSE values of each model

y_axis = [rmse(Y_test, lasso.predict(X_test)), rmse(Y_test, ridge.predict(X_test)), rmse(Y_test, svr.predict(X_test_scale))]
x_axis = ['Lasso', 'Ridge', 'Support Vector']

plt.bar(x_axis, y_axis)
plt.title('RMSE values of each Model')
plt.xlabel('Models')
plt.ylabel('RMSE values')
plt.show()
```
